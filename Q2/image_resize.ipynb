```python
import cv2
%matplotlib inline
#The line above is necesary to show Matplotlib's plots inside a Jupyter Notebook
from matplotlib import pyplot as plt
import os
```

```python
dir_input = 'data/downloadedUnsplash/'
dir_output = 'data/unsplashResized/'
dir_output_ct = 'data/color_transfer_image/'
```

```python
for filename in os.listdir(dir_input):
    input_image =cv2.imread(dir_input+filename)
    resizedimg = cv2.resize(input_image, (512,512), interpolation = cv2.INTER_AREA)
    plt.imshow(resizedimg)
    plt.show()
    print(dir_output+filename)
    cv2.imwrite(dir_output+filename, resizedimg)
```

```python
for filename in os.listdir(dir_output_ct):
    input_image =cv2.imread(dir_output_ct+filename)
    plt.imshow(input_image)
    plt.show()
```
